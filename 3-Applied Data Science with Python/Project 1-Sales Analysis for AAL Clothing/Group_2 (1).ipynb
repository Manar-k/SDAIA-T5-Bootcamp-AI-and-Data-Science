```python
# 1-C Rename Cols & formatting
# 1-D Data Type check
print(sales.dtypes)
```

```python
# convert date
sales['Date'] = pd.to_datetime(sales['Date'], format='%d-%b-%Y')
# convert time
sales['Time'] = sales['Time'].astype('string')
# convert State
sales['State'] = sales['State'].astype('string')
# convert Group
sales['Group'] = sales['Group'].astype('string')

sales.head()
```

```python
# after convert
sales.dtypes
```

## 2. Categorical data cleaning:

```python
# 2-A Check format different level to be all in same lower or upper case
def clean_check(df_categor):
  print('is alpha: ',df_categor.str.isalpha().value_counts())
  print('is lower: ',df_categor.str.islower().value_counts(),"\n")

clean_check(sales['Time'])
clean_check(sales['State'])
clean_check(sales['Group'])
```

```python
# after clean
def clean_Categorical(df_categor):
  return df_categor.str.strip().str.lower()

sales['Time'] = clean_Categorical(sales['Time'])
sales['State'] = clean_Categorical(sales['State'])
sales['Group'] = clean_Categorical(sales['Group'])
```

```python
# check if Categorical is clean
clean_check(sales['Time'])
clean_check(sales['State'])
clean_check(sales['Group'])
```

```python
# 2-B Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

dummy_df = pd.DataFrame()

for column in ['Date', 'Time', 'State', 'Group']:
    dummy_df[column]  = le.fit_transform(sales[column])

dummy_df['Unit']  = sales['Unit']
dummy_df['Sales']  = sales['Sales']
dummy_df
```

```python
# 2-C Missing
sales['State'].fillna(sales['State'].mode().iloc[0], inplace=True)
display(sales)
```

## 3. Numeric data cleaning:

```python
# 3-A Missing
sales['Sales'].fillna(sales['Sales'].mean, inplace=True)
display(sales)
```

```python
# 3-B Check Logical values
```

```python
# 3-C Normalization
for column in ['Unit', 'Sales']:
    dummy_df[column] = dummy_df[column]  / dummy_df[column].abs().max()

display(dummy_df)
```

# Data Analysis

## 1. Perform descriptive statistical analysis on the sales data, including mean, median, mode, range, quartilers, and standard deviation.


```python
def descriptive_analysis(df):
    info = df.describe()
    print()
    for mode in df.mode():
        info['mode'] = mode
    info['range'] = info['max'] - info['min']
    print(info)
```

```python
descriptive_analysis(sales['Unit'])
```

```python
descriptive_analysis(sales['Sales'])
```

```python
print(sales["Time"].value_counts())
print(sales["State"].value_counts())
print(sales["Group"].value_counts())
print(sales["Date"].value_counts())
```

## 2. Identify the groups and states that are generating the highest and lowest sales.

```python
group_sales = sales.groupby('Group')['Sales'].sum()
highest_group = group_sales[group_sales==group_sales.max()]
lowest_group = group_sales[group_sales==group_sales.min()]

print(f"Highest sales group:\n", highest_group)
print(f"Lowest sales group:\n", lowest_group)
```

```python
group_sales = sales.groupby('State')['Sales'].sum()
highest_group = group_sales[group_sales==group_sales.max()]
lowest_group = group_sales[group_sales==group_sales.min()]

print(f"Highest sales group:\n", highest_group)
print(f"Lowest sales group:\n", lowest_group)
```

## 3. Generate weekly, monthly, and quarterly reports for the analysis made.

```python
sales['Date'] = pd.to_datetime(sales['Date'])
weekly_report = sales.resample('W', on='Date').sum(numeric_only = True)
monthly_report = sales.resample('M', on='Date').sum(numeric_only = True)
quarterly_report = sales.resample('Q', on='Date').sum(numeric_only = True)
print("Weekly Report:")
print(weekly_report)
print()
print("Monthly Report:")
print(monthly_report)
print()
print("Quarterly Report:")
print(quarterly_report)
```

# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette("pastel")
```

## 1. **State-wise sales** analysis for different groups (kids, women, men, and seniors)

## 3. **Time-of-the-day** analysis: during which time of the day are sales the highest, and during which time are sales the lowest?

```python
# sales during time of day chart
t = sales.groupby('Time')['Sales'].sum()
plt.pie(t, labels= t.index)
plt.show()
```

```python
figure, axes = plt.subplots(2,2, figsize = (35,10))
axes[0,0].plot(sales.resample('D', on='Date').sum())
axes[0,0].set_title('Daily')

axes[0,1].plot(weekly_report)
axes[0,1].set_title('Weekly')

axes[1,0].plot(monthly_report)
axes[1,0].set_title('Monthly')

axes[1,1].plot(quarterly_report)
axes[1,1].set_title('Quarterly')
plt.show()
```
